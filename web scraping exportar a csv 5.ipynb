```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#definir url
url='https://pokemondb.net/pokedex/all'
r=requests.get(url)
```

```python
#convertir en sopa
soup=bs(r.content,'html.parser')
```

```python
#buscar
rows=soup.find('table',attrs={'id':'pokedex'}).find('tbody').find_all('tr')
```

```python
#buscar
rows[0].find_all('td')[1].get_text()
```

```python
#definir las filas
names=[]
escribe=[]
total=[]
hp=[]
ataque=[]
defensa=[]
esp=[]
esp_d=[]
vel=[]
for row in rows:
    names.append(row.find_all('td')[1].get_text())
    escribe.append(row.find_all('td')[2].get_text())
    total.append(row.find_all('td')[3].get_text())
    hp.append(row.find_all('td')[4].get_text())
    ataque.append(row.find_all('td')[5].get_text())
    defensa.append(row.find_all('td')[6].get_text())
    esp.append(row.find_all('td')[7].get_text())
    esp_d.append(row.find_all('td')[8].get_text())
    vel.append(row.find_all('td')[9].get_text())
```

```python
#armar el dataframe 
pokedex_df=pd.DataFrame({'nombre':names,
                         'tipo':escribe,
                         'total':total,
                         'hp':hp,
                         'ataque':ataque,
                         'defensa':defensa,
                         'especial ataque':esp,'especial defensa':esp_d,'velocidad':vel})
```

```python
#exportar a csv
pokedex_df.to_csv('pokedex_csv.csv')
```
