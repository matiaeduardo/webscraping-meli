```python
from bs4 import BeautifulSoup as bs
import requests
from lxml import etree
```

```python
r=requests.get('https://pokemondb.net/pokedex/all')
```

```python
r.status_code
```

```python
soup=bs(r.content,'html.parser')
```

```python
dom=etree.HTML(str(soup))#para realizar la busqueda con xpath
```

```python
name=dom.xpath("//table[@id='pokedex']/tbody/tr/td[@class='cell-name']/a")
```

```python
names=[i.text for i in name]
```
